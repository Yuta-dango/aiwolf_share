# 全てをembedding

```python
Agents = ["Agent[00]", "Agent[01]", "Agent[02]", "Agent[03]", "Agent[04]"]
Roles = ["VILLAGER", "SEER", "MEDIUM", "BODYGUARD", "WEREWOLF", "POSSESSED"]
Teams = ["BLACK", "WHITE"]
```

```python
datalist = []
for column in df.columns: 
    original_sentences = df[column].dropna().values 
    for original_sentence in original_sentences:
        if "[AGENT]" in original_sentence and "[ROLE]" in original_sentence:
            for x in Agents:
                for y in Roles:
                    ans = {} # 各sentenceの情報を入れる辞書
                    ans["sentence"] = original_sentence.replace('[AGENT]', x).replace('[ROLE]', y)
                    ans["Agent"] = x
                    ans["Role"] = y
                    ans["Team"] = None
                    datalist.append(ans)
        elif "[AGENT]" in original_sentence and "[TEAM]" in original_sentence:
            for x in Agents:
                for y in Teams:
                    ans = {} # 各sentenceの情報を入れる辞書
                    ans["sentence"] = original_sentence.replace('[AGENT]', x).replace('[TEAM]', y)
                    ans["Agent"] = x
                    ans["Role"] = None
                    ans["Team"] = y
                    datalist.append(ans)
        elif "[AGENT]" in original_sentence:
            for x in Agents:
                ans = {} # 各sentenceの情報を入れる辞書
                ans["sentence"] = original_sentence.replace('[AGENT]', x)
                ans["Agent"] = x
                ans["Role"] = None
                ans["Team"] = None
                datalist.append(ans)
        elif '[ROLE]' in original_sentence:
            for y in Roles:
                ans = {} # 各sentenceの情報を入れる辞書
                ans["sentence"] = original_sentence.replace('[ROLE]', y)
                ans["Agent"] = None
                ans["Role"] = y
                ans["Team"] = None
                datalist.append(ans)
```

```python
Df_all = pd.DataFrame(datalist)
Df_all
```

```python
Df_all[Df_all["sentence"].str.contains('[TEAM]')]
```

```python
from embedding import get_embedding
Df_all["embedded"] = Df_all["sentence"].map(lambda x: get_embedding(x))
```

```python
Df_all.to_csv("new_embedded_all.csv")
```
